## Cleaning Data
In this section, clean **all** of the issues you documented while assessing.

```python
twitter_archive_c = twitter_archive.copy()
image_predictions_c = image_predictions.copy()
additional_c = additional.copy()
```

### Issue #1: (Tidiness)

#### Define: working on both tables (`twitter_archive_c` and `additional_c`) individually takes a lot of time. So we will merge them

#### Code

```python
additional_c["tweet_id"] = additional_c["id"]#Changing the name of the id column in 'additional_c' to match 'twitter_archive_c'
additional_c.drop("id", axis = 1, inplace = True)
twitter_archive_c = pd.merge(twitter_archive_c, additional_c, on="tweet_id", how="inner")
```

### Issue #6: (Tidiness)

### Issue #7: (Quality)

#### Define: no data in column dog_state(s). So let's add them

#### Code

```python
#We need a function to detect the dog state for each row
def detect_dog_state(i):
    if "doggo" in i:
        return "doggo"
    elif "pupper" in i:
        return "pupper"
    elif "puppo" in i:
        return "puppo"
    elif "floof" in i:
        return "floof"
    else:
        return "None"

twitter_archive_c['dog_state(s)'] = twitter_archive_c['text'].apply(detect_dog_state) #apply the function
```

#### Test

```python
twitter_archive_c['dog_state(s)'].value_counts()
```

### Issue #8: (Quality)

#### Define: Names of dogs are incorrectly set. So, let's set them right

#### Code

```python
names = twitter_archive["name"].unique()
for j in twitter_archive_c["name"]:
    for i in names:
        if i in twitter_archive["text"]:
            j = i
        else:
            j = "Unknown name"
#Now the series needs to be filtered:
```

```python
bad_names = ['None', 'a', 'the', 'an', 'this', 'very', 'my', 'his', 'just', 
             'not', 'one', 'all', 'incredibly', 'quite', 'by', 
             'getting', 'actually', 'officially', 'life', 'light', 'old']
def filterer(i):
    if i in bad_names:
        return "Unknown name"
    return i
twitter_archive_c["name"] = twitter_archive_c["name"].apply(filterer) #Now it is ready
```

#### Test

```python
twitter_archive_c["name"].value_counts()
```

### Issue #10: (Tidiness)

#### Define:  the columns "rating_denominator" and "rating_numerator" should be in one column "ratings_out_of_ten"

```python
twitter_archive_c.insert(twitter_archive_c.columns.get_loc("rating_numerator"), "ratings_out_of_10" , 0) #keep it empty for now
```
